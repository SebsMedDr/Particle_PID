```python
import pid_df
import matplotlib.pyplot as plt
from seaborn import distplot,jointplot

pid_pr_geant3 = pid_df.pid_df('../SingleTracks/tracks_geant3_pr.csv','pr')
pid_pr_geant4 = pid_df.pid_df('../SingleTracks/tracks_geant4_pr.csv','pr')
pid_pr_old_geant3 = pid_df.pid_df('../MyDanielFiles/ProtDF_sh_DataFrame.csv','pr')
```

```python
p_bool_geant3 = pid_pr_geant3.pid.p < 8.0
theta_bool_geant3 = pid_pr_geant3.pid.theta < 3.14/2
bool_geant3 = p_bool_geant3 & theta_bool_geant3
```

```python
distplot(pid_pr_geant3.pid[pid_pr_geant3.pid.p < 20.0].p,bins=200,kde=False,color='k',hist_kws={'alpha':1})
plt.xlim(0,20)
#plt.yscale('log')
plt.xlabel('Momentum (GeV/c)')
plt.ylabel('Counts')
```

```python
distplot(pid_pr_geant3.pid[bool_geant3].p,bins=100,kde=False,color='k',hist_kws={'alpha':1})
plt.xlim(0,10)
#plt.yscale('log')
plt.xlabel('Momentum (GeV/c)')
plt.ylabel('Counts')
plt.savefig('p_geant3.pdf')
```

```python
distplot(pid_pr_old_geant3.pid.p,bins=100,kde=False,color='k',hist_kws={'alpha':1})
plt.xlim(0,10)
#plt.yscale('log')
plt.xlabel('Momentum (GeV/c)')
plt.ylabel('Counts')
plt.savefig('p_old_geant3.pdf')
```
