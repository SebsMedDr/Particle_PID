```python
import pid_df
import matplotlib.pyplot as plt

pid_ep = pid_df.pid_df('EpDF_DataFrame.csv','ep',0.1)
pid_ep.add_df('PipDF_DataFrame.csv','pip',0.02)

feature_list = ['E1E9_FCAL','E9E25_FCAL','dE_BCAL','dE_BCAL_L2','dE_BCAL_L3','dE_BCAL_L4','dE_BCAL_preshower','dE_FCAL','dEdx_ST','p','sigLong_BCAL','sigTrans_BCAL','sumU_FCAL','sumV_FCAL','theta']
pid_ep.set_features(feature_list,'particle_id')

pid_ep.nn_model('mms')

pid_ep.predict()
pid_ep.roc_auc()
fig,ax = plt.subplots(figsize=(20,15))
pid_ep.roc_curve()
print(pid_ep.lr_auc)
print(pid_ep.accuracy)
fig.savefig('nn_mms.pdf')
```

```python
pid_ep.restore_df()
pid_ep.nn_model('ss')

pid_ep.predict()
pid_ep.roc_auc()
fig,ax = plt.subplots(figsize=(20,15))
pid_ep.roc_curve()
print(pid_ep.lr_auc)
print(pid_ep.accuracy)
fig.savefig('nn_ss.pdf')
```

```python
pid_ep.k_fold_cv(10)
count = 0
for train,test in pid_ep.pid_k_fold_split:
    if count == 0:
        print(len(train),len(test))
    pid_ep.nn_model_cv(train,test)
    count+=1
```

```python
import matplotlib.pyplot as plt
from statistics import mean,variance,stdev

plt.scatter(range(1,len(pid_ep.accuracies)+1),pid_ep.accuracies)
plt.xlabel('K-fold')
plt.ylabel('Accuracy')

print(mean(pid_ep.accuracies))
print(stdev(pid_ep.accuracies))
plt.savefig('k-fold_nn_first_pass_accuracies.pdf')
```

```python
plt.scatter(range(1,len(pid_ep.lr_aucs)+1),pid_ep.lr_aucs)
plt.xlabel('K-fold')
plt.ylabel('AUC')

print(mean(pid_ep.lr_aucs))
print(stdev(pid_ep.lr_aucs))
plt.savefig('k-fold_nn_first_pass_lr_aucs.pdf')
```
